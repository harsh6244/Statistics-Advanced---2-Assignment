**Q.3 Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test**

Answer - The significance level, commonly denoted by (alpha), is a crucial concept in hypothesis testing. It helps you decide whether to reject or fail to reject the null hypothesis (H₀) based on the probability of observing your test results under the assumption that the null hypothesis is true.

**Role in Hypothesis Testing**

*   Set Before Testing:
Before performing a test, the researcher selects a significance level (e.g., α = 0.05).

*   Compare with p-value:
After calculating the p-value (the probability of observing the data assuming H₀ is true), you:

      Reject H₀ if: p-value ≤ α

     Fail to reject H₀ if: p-value > α

*   Determines Critical Region:
The significance level helps define the critical region — the set of values where H₀ is rejected.

**Q.6 Write a Python program to generate a binomial distribution with n=10 and

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10        # Number of trials
p = 0.5       # Probability of success
size = 1000   # Number of samples

# Generate binomial distribution data
data = np.random.binomial(n=n, p=p, size=size)

# Plot histogram
plt.hist(data, bins=range(n + 2), edgecolor='black', align='left')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
**Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]

# Hypothesized population mean
mu_0 = 50

# Population standard deviation (assumed known)
sigma = 0.5

# Sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Z-statistic calculation
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Significance level
alpha = 0.05

# Output results
print(f"Sample Mean = {sample_mean:.4f}")
print(f"Z-statistic = {z:.4f}")
print(f"P-value = {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: Significant difference from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from population mean.")
```

**Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
mean = 100
std_dev = 15
sample_size = 100

data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

# Calculate the 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use sample standard deviation
confidence_level = 0.95
alpha = 1 - confidence_level

# Get the t-critical value (since sigma is unknown)
t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)

# Margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 3: Plot the histogram with confidence interval
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle='--', label=f'95% CI Lower = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='--', label=f'95% CI Upper = {ci_upper:.2f}')
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

**Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.**


```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = [(x - mean) / std_dev for x in data]

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z = 0)')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print a few sample Z-scores
    print("Sample Z-scores (first 5):", z_scores[:5])
    return z_scores

# Example usage with a sample dataset
sample_data = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]
z_scores = calculate_and_plot_z_scores(sample_data)
```
